```python
from tkinter import *
from openpyxl.workbook import Workbook
from openpyxl import load_workbook
from tkinter import messagebox as msgbox
import sqlite3

def stockdb():
    stock=Tk()
    
    stock.geometry("800x750")
    stock.title("stock database")
    def quit():
        stock.destroy()
        
    home=Button(stock,text="HOME",width=10,bg="purple2",command=quit).place(x=650,y=45)
    head=Label(stock,text="STOCK DATABASE",font=("bold",20),bg="orange2").place(x=250,y=10)
    id=Label(stock,text="Product ID",font=("bold",10)).place(x=20,y=60)
    name=Label(stock,text="Product name",font=("bold",10)).place(x=20,y=90)
    quan=Label(stock,text="Quantity",font=("bold",10)).place(x=20,y=120)
    cost=Label(stock,text="Cost",font=("bold",10)).place(x=20,y=150)
    
    def update():
        file=load_workbook("stock.xlsx")
        sheet=file.active
        sheet.cell(column=1,row=sheet.max_row+1,value=idd.get())
        sheet.cell(column=2,row=sheet.max_row,value=na.get())
        sheet.cell(column=3,row=sheet.max_row,value=qua.get())
        sheet.cell(column=4,row=sheet.max_row,value=pri.get())
        file.save("stock.xlsx")
        mgg=msgbox.showinfo("update","product added")
        
    def get_products():
        wb=Workbook()
        wb=load_workbook("stock.xlsx")
        ws=wb.active
        column_a=ws['A']
        column_b=ws['B']
        column_c=ws['C']
        column_d=ws['D']
        list_a=''
        list_b=''
        list_c=''
        list_d=''
        for cell in column_a:
            list_a=f'{list_a+str(cell.value)}\n'
        iddd.config(text=list_a)
        for cell in column_b:
            list_b=f'{list_b+str(cell.value)}\n'
        name.config(text=list_b)
        for cell in column_c:
            list_c=f'{list_c+str(cell.value)}\n'
        quan.config(text=list_c)
        for cell in column_d:
            list_d=f'{list_d+str(cell.value)}\n'
        cost.config(text=list_d)   
        
    idd=Entry(stock,width=20)
    idd.place(x=150,y=60)
    na=Entry(stock,width=20)
    na.place(x=150,y=90)
    qua=Entry(stock,width=20)
    qua.place(x=150,y=120)
    pri=Entry(stock,width=20)
    pri.place(x=150,y=150)
    
              
    search=Button(stock,text="Search",width=10,font=("bold",10),bg="pale violet red").place(x=350,y=60)
    add=Button(stock,text="Add product",font=("bold",10),width=15,bg="medium sea green",command=update).place(x=150,y=200)
    view=Button(stock,text="View products",width=15,bg="dark khaki",command=get_products).place(x=60,y=350)
    update=Button(stock,text="Update",width=15,bg="hot pink").place(x=240,y=350)
    prin=Button(stock,text="print",width=15,bg="salmon").place(x=420,y=350)
    sf=Button(stock,text="sort",width=15,bg="olivedrab").place(x=600,y=350)
    
    iddd=Label(stock,width=20,height=20)
    iddd.place(x=10,y=380)
    name=Label(stock,width=25,height=20)
    name.place(x=120,y=380)
    quan=Label(stock,width=20,height=20)
    quan.place(x=280,y=380)
    cost=Label(stock,width=20,height=20)
    cost.place(x=410,y=380)
    
    stock.mainloop()
               
stockdb()

    
```
