```python
from tkinter import *
from openpyxl.workbook import Workbook
from openpyxl import load_workbook
from tkinter import messagebox as msgbox

def custdb():
    cust=Tk()
    cust.geometry("900x800")
    cust.title("customers")
    def quit():
        cust.destroy()
    home=Button(cust,text="HOME",width=10,bg="purple2",command=quit).place(x=650,y=45)
    head=Label(cust,text="CUSTOMER DATABASE",font=("bold",20),bg="orange2").place(x=250,y=10)
    nam=Label(cust,text="NAME",font=("bold",10)).place(x=20,y=60)
    bna=Label(cust,text="BUSINESS NAME",font=("bold",10)).place(x=20,y=90)
    gst=Label(cust,text="GSTIN",font=("bold",10)).place(x=20,y=120)
    city=Label(cust,text="CITY",font=("bold",10)).place(x=20,y=150)
    billd=Label(cust,text="BILL DUE",font=("bold",10)).place(x=20,y=180)
    def add_cust():
        file=load_workbook("customers.xlsx")
        sheet=file.active
        sheet.cell(column=1,row=sheet.max_row+1,value=idd.get())
        sheet.cell(column=2,row=sheet.max_row,value=na.get())
        sheet.cell(column=3,row=sheet.max_row,value=qua.get())
        sheet.cell(column=4,row=sheet.max_row,value=pri.get())
        sheet.cell(column=5,row=sheet.max_row,value=du.get())
        file.save("customers.xlsx")
        mgg=msgbox.showinfo("update","New customer added")
        
    add=Button(cust,text="Add",font=("bold",10),width=8,bg="medium sea green",command=add_cust).place(x=400,y=180)
    idd=Entry(cust,width=25)
    idd.place(x=150,y=60)
    na=Entry(cust,width=25)
    na.place(x=150,y=90)
    qua=Entry(cust,width=25)
    qua.place(x=150,y=120)
    pri=Entry(cust,width=25)
    pri.place(x=150,y=150)
    du=Entry(cust,width=25)
    du.place(x=150,y=180)
    
    def view_cust():
        wb=Workbook()
        wb=load_workbook("customers.xlsx")
        ws=wb.active
        column_a=ws['A']
        column_b=ws['B']
        column_c=ws['C']
        column_d=ws['D']
        column_e=ws['E']
        list_a=''
        list_b=''
        list_c=''
        list_d=''
        list_e=''
        for cell in column_a:
            list_a=f'{list_a+str(cell.value)}\n'
        iddd.config(text=list_a)
        for cell in column_b:
            list_b=f'{list_b+str(cell.value)}\n'
        name.config(text=list_b)
        for cell in column_c:
            list_c=f'{list_c+str(cell.value)}\n'
        quan.config(text=list_c)
        for cell in column_d:
            list_d=f'{list_d+str(cell.value)}\n'
        cost.config(text=list_d)
        for cell in column_e:
            list_e=f'{list_e+str(cell.value)}\n'
        due.config(text=list_e)
                    
    
    view=Button(cust,text="View",width=15,bg="dark khaki",command=view_cust).place(x=40,y=250)
    update=Button(cust,text="Update",width=15,bg="hot pink").place(x=260,y=250)
    prin=Button(cust,text="print",width=15,bg="salmon").place(x=460,y=250)

    
    iddd=Label(cust,width=20,height=20)
    iddd.place(x=5,y=280)
    name=Label(cust,width=25,height=20)
    name.place(x=110,y=280)
    quan=Label(cust,width=35,height=20)
    quan.place(x=250,y=280)
    cost=Label(cust,width=25,height=20)
    cost.place(x=450,y=280)
    due=Label(cust,width=15,height=20)
    due.place(x=620,y=280)
    cust.mainloop()
custdb()
```
